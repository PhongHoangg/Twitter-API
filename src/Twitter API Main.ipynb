```python
df = pd.read_csv('County_Centers.csv', encoding= 'unicode_escape')  
```

```python
df
```

```python
def get_tweet_data(loc):
    df_coord_county = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'geocode:"{}"'.format(loc)).get_items(), 50))
    return df_coord_county
```

```python
sample_df_ohio = pd.DataFrame()
for index, values in df.iterrows():
    loc = str(values.latitude) + ',' + str(values.longitude) + ',' + str(values.radius_km) + 'km'
    df_coord_county = get_tweet_data(loc)
    sample_df_ohio = pd.concat([sample_df_ohio, df_coord_county])
```
